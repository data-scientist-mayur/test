```python
# from sqlite3 import connect
from django.db import connections

#important libraries 
import pandas as pd
import numpy as np

pd. set_option("display.max_rows", None)
pd.set_option('display.max_columns', None)

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

from webdriver_manager.chrome import ChromeDriverManager

from fuzzywuzzy import fuzz
from fuzzywuzzy import process


import os
import pymysql
import pandas as pd

host = "127.0.0.1"
port = 3316
user = "root"
password =""
database = "open_db"

conn = pymysql.connect(
    host=host,
    port=int(3316),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

df1 = pd.read_sql_query("select email from users where  date(created_at) between '2020-01-01' and '2020-01-01' AND EMAIL NOT LIKE '%phased_out%' ",conn)
df1 = df1.head(14)
#df1

df2 = pd.read_sql_query("select email from users where  date(created_at) between '2020-01-11' and '2020-01-20' AND EMAIL NOT LIKE '%phased_out%' ",conn)
df2 = df2.head(14)
#df2


df3 = pd.read_sql_query("select email from users where  date(created_at) between '2020-01-01' and '2020-01-30' AND EMAIL NOT LIKE '%phased_out%' ",conn)
df3= df3.head(14)
#df3

#row-bind two DataFrames and reset index values
df55= pd.concat([df1,df2,df3]).reset_index(drop=True)
#print(df55)
df66 = df55.email


names=pd.DataFrame(columns=["score","email","string","head","link"])
for i in df66:
    from selenium import webdriver
    inner=[]
    options = webdriver.ChromeOptions()
    options.headless = True


    driver = webdriver.Chrome(ChromeDriverManager().install(),options = options)
    driver.maximize_window()
    driver.get("https://www.google.co.in/")
    search_box = driver.find_element_by_name("q")
    search_box.send_keys( i)
    search_box.submit()
    elements = driver.find_elements_by_css_selector(".lEBKkf")
    head=driver.find_elements_by_css_selector(".yuRUbf")
    link = driver.find_elements_by_css_selector(".tjvcx")
    
    storyTitles = [el.text for el in elements]
    storyTit = [et.text for et in head]
    storylink = [em.text for em in link]
    innerHead=[]
    innerLink=[]
    for el in elements:
        inner.append(el.text)
    for et in head:
        innerHead.append(et.text)
    for em in link:
        innerLink.append(em.text)
    if process.extractOne(i, inner)  != None:    
        score=process.extractOne(i, inner)[1] 
        string=process.extractOne(i, inner)[0] 
    #    print(process.extractOne(i, inner))
        
        for index,val in enumerate(elements):
            if string==val.text:
                head=innerHead[index]
                link=innerLink[index]
        names=names.append({'score':score,"email":i,"string":string,"head":head,"link":link},ignore_index=True) 
    else:
        names=names.append({'score':0,"email":i,"string":None,"head":None,"link":None},ignore_index=True) 
    driver.close()
    
    
df11 =names[(names['score'] >= 90)]

df11.to_html("11april_scrap.html")



```
